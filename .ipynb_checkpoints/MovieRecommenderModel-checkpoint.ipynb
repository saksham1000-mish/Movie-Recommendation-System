# MOVIE RECOMMNEDATION SYSTEM (Content-Based)

```python
import numpy as np   # linear algebra
import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)
import ast
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
#read data from the csv file
movies_df = pd.read_csv('tmdb_5000_movies.csv')
credits_df = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies_df.info()
```

```python
credits_df.head(2)
```

```python
credits_df.shape
```

```python
#merging the 2 dataframes on the basis of the title of the movie
newmovies = movies_df.merge(credits_df,on='title')
```

```python
newmovies.head(2)
```

```python
#for content-based recommender we need only few of the columns
movies = newmovies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(3)
```

merge overview, genres, keywords, cast in a new column tags
to do that...first we will do pre-processing(modify the existing columns, remove duplicates, remove nill items,etc)

We will start by **preprocessing the data** <br>
>First remove nill and duplicate items. <br>
>We will be merging columns (overview, genres, keywords, cast and crew) into a single column as **'tags'**. <br>

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
covert this into -->
['action','adventure','fantasy,'scifi']

```python
#we need only the name of the genre and not it's id.
def convert(obj):
    list=[]
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
#same thing is to be done with the keywords column
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

for the cast column we will be picking just top 3 actors

```python
#now for the cast column we will only pick the top 3 actors
def convertT(obj):
    list=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            list.append(i['name'])
            counter+=1
        else:
            break
    return list
```

```python
movies['cast'] = movies['cast'].apply(convertT)
```

```python
#Out of all the values in crew, we will be picking only the directoer value and nothing else
movies['crew'][0]
```

```python
def director(obj):
    list=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

Now a problem may arise where the function might treat the first and the last name of a person as 2 separate enitities<br>
Hence, we need to remove spaces between the names like :- <br>
<t>Gore Verbinski ---> GoreVerbinski

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
#Concatenating all the columns in a single column - 'tags'
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords']+ movies['cast']+movies['crew']
new_df = movies[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:' '.join(x))
```

Yet another problem arises where there are words like action and actions which are same but form separate features.<br>
Hence we need to tackle this problem with a library known as PorterStemmer

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

We will be applying a **Text Vectorisation** techinque known as **Bag of Words**.
> concatenate all tags(big string created)
> lets say we extract 5000 most recurring words 
> now we will see how much time those words appear in eachmovie tag. a table of (5000,5000) will be created and a vector for each movie will be formed 

**NOTE** - We wont be using stop words(are and of is the, etc) and remove them with the help of a SCIKIT - CountVectorizer

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

now we need to calculate vector distance of every movie with every movie
we wont calc the eucledian dist(r1-r2)..we will calculate cosine distance(angle b/w them)
function in sklenarn --> cosine similarity

```python
#cosine similarity is used which gives the angle between the 2 vectors and not its eucledian distance.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity[0]
#this gives the similarity of the 1st movie with every other movie
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
#we use enumerate so as not to lose the index position when we sort the similairty
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Batman Begins')
```
